```python
def get_IA_perc_ZIP(url_part):
    #scrapes table from http://zipatlas.com/us/ia/zip-code-comparison/population-density.htm 
    #and http://zipatlas.com/us/ia/zip-code-comparison/population-density.?.htm
    #Define websites
    url_base = "http://zipatlas.com/us/ia/zip-code-comparison/"
    page1 = "percentage-" + url_part + "-population"+ "."
    pnum = [i for i in range(1,11)]
    htm = "htm"
    tables_pages =[]
    #Pull information from each of the 10 subpages (fixed to 10 as per website format)
    for page in pnum:
        #The first page is different
        if page==1:
            url = url_base + page1+ htm
        else:
            url = url_base + page1 + str(page) + "." + htm
        r = requests.get(url)
        soup = BeautifulSoup(r.text)
        #filtering on rules: all after inspecting table; it returns unique result
        tables = soup.find_all('table', {'rules': 'all'})
        #Nice piece of code taken from someone else
        tables_pages.append([[c.string for c in row.findAll("td")] for row in tables[0].findAll('tr')])
    #Output is list of length 10 (1 for each page); each of these lists contains 101 lists (one for each table row)
    #Define headers
    headers=tables_pages[0][0]
    df = pd.DataFrame(columns=[i+"_"+url_part for i in headers])
    for j in range(len(tables_pages)):
        for i in range(len(tables_pages[j])):
            ji = j*100+i
            if tables_pages[j][i]!=headers:
                df.loc[ji] = tables_pages[j][i]
    return df
```
